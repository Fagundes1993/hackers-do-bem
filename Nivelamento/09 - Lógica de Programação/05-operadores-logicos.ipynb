# Operador AND

# Operador OR

```python
entrada_1 = False
entrada_2 = False

resultado = entrada_1 or entrada_2

print(resultado)
```

```python
entrada_1 = True
entrada_2 = False

resultado = entrada_1 or entrada_2

print(resultado)
```

```python
entrada_1 = False
entrada_2 = True

resultado = entrada_1 or entrada_2

print(resultado)
```

```python
entrada_1 = True
entrada_2 = True

resultado = entrada_1 or entrada_2

print(resultado)
```

## Exemplo de operação relacional com operação lógica

```python
(((2*2) == 4) or ((16/4) == 4) or (5*20) != 75)
```

```python
(((2*2) != 4) or ((16/4) != 4) or (5*20) == 75)
```

# Operador NOT

```python
entrada = True
resultado = not entrada

print(resultado)
```

```python
entrada = False
resultado = not entrada

print(resultado)
```
