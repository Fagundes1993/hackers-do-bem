## Exemplo 1

```python
idade = int(input("Digite a sua idade:\n"))
if(idade >=18):
  print("A pessoa é maior de idade")
print("Fim de programa")
```

## Exemplo 2

```python
idade = int(input("Digite a sua idade:\n"))
if(idade >=18):
  print("A pessoa é maior de idade")
if(idade <18):
  print("A pessoa é menor de idade")
print("Fim de programa")
```
