# Curso Hackers do Bem - Aula 3 - Operadores Relacionais

Exemplo: Faça um programa que tenha a seguinte sequencia de atividades.

1 - receba um valor de um usuário e armazene na variável num

2 - Receba outro valor e armazena na variavel comparação

3 -  Converta ambos os valores para inteiro

4 - Realize a operação resultado = 2 * num + 3

5 - Utilize operadores lógicos para comparar se o resultado é maior ou igual a comparação

```python
#entrada de dados
num = int(input("Digite um numero: "))
comparacao = int(input("Digite outro numero: "))

#processamento de dados
resultado = 2 * num + 3

print(f"O resultado da comparação é: {resultado >= comparacao}")
```
