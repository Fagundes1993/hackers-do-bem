## Cálculo do IMC

```python
# Entrada de dados

altura = input("Digite sua altura: ")
peso = input("Digire o seu peso: ")
idade = input("Digite sua idade: ")

# Processamento

altura = float(altura)
peso = float(peso)
idade = int(idade)

imc = peso/(altura*altura)
print(imc)
```

### Versão Aprimorada do IMC

```python
# Entrada de dados

altura = float(input("Digite sua altura: "))
peso = float(input("Digire o seu peso: "))
idade = int(input("Digite sua idade: "))

# Processamento

imc = peso/(altura*altura)
print(f"O seu IMC é igual a: {imc:.2f} e sua idade é igual a {idade}")
```
