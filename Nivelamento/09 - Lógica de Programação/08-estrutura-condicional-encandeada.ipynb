Conversão de notas para conceito

Nota          Conceito

n >= 9            A

9 > n >= 7        B

7 > n >= 6        C

6 > n >= 5        D

n < 5             F

```python
# Entrada de dados

nota = float(input("Informe a sua nota: \n"))

#Processamento e Saída
if(nota >= 9):
  print("Nota conceito: A")
elif(nota >=7):
  print("Nota conceito: B")
elif(nota >= 6):
  print("Nota conceito: C")
elif(nota >=5):
  print("Nota conceito: D")
else:
  print("Nota conceito: F")

print("FIM DO PROGRAMA")
```
