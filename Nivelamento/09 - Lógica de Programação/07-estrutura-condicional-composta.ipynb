## Exemplo 1

```python
# Entrada de dados

media = int(input("informe a média do aluno: "))

# processamento e saida e dados

if (media >= 50):
  print("O aluno foi aprovado!")
else:
  print("O aluno foi reprovado.")

print("Fim do programa.")
```
