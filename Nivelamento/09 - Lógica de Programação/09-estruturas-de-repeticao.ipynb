##Comando While

### Exemplo 1 - Vamos criar um programa que receba um número é retorne a sua tabuadad e 1 até

Sem o uso do comando while

```python
# Entrada de dados

numero = int(input("Digite um número para o cálculo de sua tabuada: \n"))

# Exibindo a tabuada
print(f"{numero} x 1 = {numero*1}")
print(f"{numero} x 2 = {numero*2}")
print(f"{numero} x 3 = {numero*3}")
print(f"{numero} x 4 = {numero*4}")
print(f"{numero} x 5 = {numero*5}")
print(f"{numero} x 6 = {numero*6}")
print(f"{numero} x 7 = {numero*7}")
print(f"{numero} x 8 = {numero*8}")
print(f"{numero} x 9 = {numero*9}")
print(f"{numero} x 10 = {numero*10}")
```

Com o uso do while

```python
# Entrada de dados
numero = int(input("Digite um número para o cálculo de sua tabuada: \n"))

#criando a variável de controle
i = 1

# Exibindo a tabuada
while(i <= 10):
  print(f"{numero} x {i} = {numero*i}")
  i = i + 1
print("FIM DE PROGRAMA")
```

### Exemplo 2 - Vamos criar um programa que cálcula a média das idaddes de 5 pessoas

Podemos ter a necessidade de somar valores a cada repetição dos laços de repetição. Para isso precisamos de uma variável acumuladora

Sem o uso do comando while

```python
#recebendo as idades
idade1 = int(input("Digite a primeira idade: \n"))
idade2 = int(input("Digite a segunda idade: \n"))
idade3 = int(input("Digite a terceira idade: \n"))
idade4 = int(input("Digite a quarta idade: \n"))
idade5 = int(input("Digite a quinta idade: \n"))

#caluculando a média:
media = (idade1 + idade2 + idade3 + idade4 + idade5)/5

#exibindo a media das idades:
print(f"A media das idades é igual a {media}")
```

Com o comando while

```python
# Criando a variável de controle do laçom contando 5 repetições
contador = 1

# Criando uma variável para acumular os valores das idades digitadas
soma = 0

# Criando o laço de repetição para receber 5 idades e soma-las
while(contador <=5):
  # Recebendo a idade do usuário
  idade = int(input(f"Digite a idade do usuário nº {contador}: "))

  # Somando a idade do usuário ao valor atual da variavel acumulador
  soma = soma + idade
  # Incrementando a variável contador para controle do laço
  contador = contador + 1

# Calculando a média das idades
media = (soma)/5

#exibindo a media das idades:
print(f"A media das idades é igual a: {media}")
print("FIM DO PROGRAMA")
```
